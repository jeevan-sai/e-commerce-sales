```python
dcols=['Country latitude', 'Country longitude','Quantity','Sales','Discount','Profit']
```

```python
for i in dcols:
    print(i)
```

```python
for i in dcols:
    skew=dfsales[i].skew().round()
    print(i,"-->",skew)
```

```python
for i in dcols:
    sns.displot(dfsales[i])
    plt.show()
    print(i)
```

```python
for i in dcols:
    skew=dfsales[i].skew().round()
    if skew !=0:
        q1=dfsales[i].quantile(0.25)
        q3=dfsales[i].quantile(0.75)
        iqr=q3-q1
        upper=q3+1.5*iqr
        lower=q1-1.5*iqr
        dfsales[i]=dfsales[i].map(lambda x:  upper if x>upper else  lower if x<lower else x)
    else :
        mean=dfsales[i].mean()
        std=dfsales[i].std()
        l=mean-3*std
        u=mean+3*std
        dfsales[i]=dfsales[i].map(lambda x:  u if x>u else l if x<l else x)
    
```
